```python
CompanyName=dataset['CarName'].apply(lambda x : x.split(' ')[0])
dataset.insert(3,"CompanyName",CompanyName)
dataset.drop(['CarName'],axis=1,inplace=True)
dataset.head()
```

```python
#fixing invalid value
dataset.CompanyName = dataset.CompanyName.str.lower()

def replace_name(a,b):
    dataset.CompanyName.replace(a,b,inplace=True)

replace_name('maxda','mazda')
replace_name('porcshce','porsche')
replace_name('toyouta','toyota')
replace_name('vokswagen','volkswagen')
replace_name('vw','volkswagen')

dataset.CompanyName.unique()
```

```python
dataset['fueleconomy'] = (0.55 * dataset['citympg']) + (0.45 * dataset['highwaympg'])
```

```python
#Binning the Car Companies based on avg prices of each Company.['price'] = cars['price'].astype('int')
temp = dataset.copy()
table = temp.groupby(['CompanyName'])['price'].mean()
temp = temp.merge(table.reset_index(), how='left',on='CompanyName')
bins = [0,10000,20000,40000]
cars_bin=['Budget','Medium','Highend']
dataset['carsrange'] = pd.cut(temp['price_y'],bins,right=False,labels=cars_bin)
dataset.head()
```

```python
cars = dataset[['price', 'fueltype', 'aspiration','carbody', 'drivewheel','wheelbase',
                  'curbweight', 'enginetype', 'cylindernumber', 'enginesize', 'boreratio','horsepower', 
                    'fueleconomy', 'carlength','carwidth', 'carsrange']]
cars.head()
```

```python
sns.pairplot(cars)
plt.show()
```

```python
# Defining the map function
def dummies(x,df):
    temp = pd.get_dummies(df[x], drop_first = True)
    df = pd.concat([df, temp], axis = 1)
    df.drop([x], axis = 1, inplace = True)
    return df
# Applying the function to the cars_lr
cars = dummies('fueltype',cars)
cars= dummies('aspiration',cars)
cars= dummies('carbody',cars)
cars= dummies('drivewheel',cars)
cars= dummies('enginetype',cars)
cars= dummies('cylindernumber',cars)
cars= dummies('carsrange',cars)
```

```python
cars.head()
```

```python
cars.shape
```

```python
from sklearn.model_selection import train_test_split
import numpy as np
np.random.seed(0)
df_train, df_test = train_test_split(cars, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
num_vars = ['wheelbase', 'curbweight', 'enginesize', 'boreratio', 'horsepower','fueleconomy','carlength','carwidth','price']
df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

```python
#Dividing data into X and y variables
y_train = df_train.pop('price')
X_train = df_train
```

Model Building

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest= train_test_split(X_train,y_train,test_size=0.1,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(xtrain,ytrain)
```

```python
pred=lm.predict(xtrain)
```

```python
from sklearn.metrics import r2_score
new=r2_score(ytrain,pred)
print("The r2 score is:- ",new)
```

Visualizing the actual price vs predicted price

```python
plt.scatter(ytrain,pred)
plt.xlabel("Actual price")
plt.ylabel("Predicted price")
plt.title("Actual prices Vs Predicted prices")
plt.show()
```

Now we predict on testing data

```python
newpred=lm.predict(xtest)
```

```python
from sklearn.metrics import r2_score
new1=r2_score(ytest,newpred)
print("The r2 score of testing dataset is:- ",new1)
```
